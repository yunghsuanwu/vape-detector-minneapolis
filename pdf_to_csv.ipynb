```python
import tabula

pdf_file = tabula.read_pdf("haloreport-2024-12-03-2025-02-03-copy.pdf", pages="all")

pdf_file.to_csv("haloreport-2024-12-03-2025-02-03-copy.csv", index=False)
```

```python
import pdfplumber
import pandas as pd

def extract_table_from_pdf(pdf_path, output_csv):
    all_rows = []
    column_names = None
    page_extracted = 0

    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            table = page.extract_table()
            if not table:
                print(f'Page {page_num+1} has no table.')
                continue  # Skip if no table found on the page
            else:
                page_extracted += 1
                # Assume first row of each page is the header
                page_header = table[0]
                page_rows = table[1:]

                # Save column names only from the first page
                if column_names is None:
                    column_names = page_header

                # Skip headers in subsequent pages
                all_rows.extend(page_rows)

    print(f'Total pages extracted: {page_extracted}')

    # Create DataFrame
    df = pd.DataFrame(all_rows, columns=column_names)

    # Optionally drop rows that are clearly not data
    df.dropna(how='all', inplace=True)

    # Save to CSV
    df.to_csv(output_csv, index=False)

```

```python
# Usage
extract_table_from_pdf('haloreport-2024-12-03-2025-02-03-copy.pdf', 'haloreport-2024-12-03-2025-02-03-copy.csv')
```
